## What is Exploratory Data Analysis?


Exploratory Data Analysis (EDA) is an approach to analyzing and summarizing datasets, with the goal of discovering patterns, anomalies, and other insights that can inform further analysis and modeling efforts. 

EDA typically involves visualizing data using plots and graphs, as well as statistical summaries and tests to uncover patterns and relationships in the data. 

The goal of EDA is not to confirm preconceived notions or hypotheses, but rather to explore the data and let it lead the analysis in new and potentially unexpected directions. 

EDA is an important first step in any data analysis project, as it helps to get a sense of the structure and characteristics of the data, identify any problems or issues that need to be addressed, and inform the direction of further analysis.

### 2. Select requried columns from the dataset.

```python
selected_col = ['id', 'name', 'city', 'rating', 'rating_count', 'cost', 'cuisine']   # Selection of required columns
swiggy_data = pd.read_csv(swiggy_csv, usecols = selected_col)                        # Reading csv file with required columns
swiggy_data
```

### 3. Viewing basic information about rows, columns and data types from the dataset.

```python
swiggy_data.info()
```

The selected column datasets contains 148,541 rows & 7 columns.

### 4. Handling missing and duplicate data.

#### Check if any column contains NA value.

```python
swiggy_data.isna().sum()
```

There are 86 data value in dataset that contains NA in `name, rating and rating_count` column. So drop columns that contain NA values.

```python
swiggy_df = swiggy_data.dropna(subset=['name', 'rating', 'rating_count']).reset_index(drop=True)
swiggy_df.isna().sum()
```

There is 45 data value in `cost` column and 13 data values in `cuisine` column that contains NA. So drop columns that contain NA values.

```python
swiggy_df = swiggy_df.dropna(subset=['cost','cuisine']).reset_index(drop=True)
swiggy_df.isna().sum()
```

There is null NA values present in `swiggy_df` dataframe

#### Check for duplicate data in `id` column.

```python
swiggy_df['id'].duplicated().sum()
```

There is no duplicate data in `id` column. Hence the id value is unique.

#### Replace '--' value to 'NaN' in `rating` column.

```python
swiggy_df.rating.replace('--', np.nan, inplace=True)
swiggy_df
```

There was `--` value in rating column so replaced it with NaN value.

### 5. Create two new columns `town and cities` from `city` column.

#### Spliting the city column data into two data values and storing it in `town & cities` column.

```python
swiggy_df[['town', 'cities']] = swiggy_df.city.str.split(',', n=1, expand=True)     
swiggy_df.cities.unique()
```

The city column contains data like `[Airoli,Mumbai], [Fort Colaba,Mumbai] , [Shivaji Nagar,Pune] , [Ghatlodia,Ahmedabad] , [BTM,Bangalore].... and etc.` 
So we have split the data and stored the splitted data into new column named as `town and cities.`

```python
swiggy_df.cities.value_counts()
```

There are two data in town columns `' New BEL Road,Bangalore' & ' Nacharam & Malkajigiri,Hyderabad'` which is inappropriate.

#### Replacing ' New BEL Road,Bangalore' with 'Bangalore' & ' Nacharam & Malkajigiri,Hyderabad' with 'Hyderabad'

```python
swiggy_df['cities'] = swiggy_df.cities.replace(to_replace = { ' New BEL Road,Bangalore' : 'Bangalore', ' Nacharam & Malkajigiri,Hyderabad' : 'Hyderabad'})
```

```python
swiggy_df.cities.value_counts()
```

### 6. Datatypes conversions of the dataset.

#### Convert `rating` column data type from object to float.

```python
swiggy_df['rating'] = swiggy_df.rating.astype(float)   # Conversion of data-type from object to float64
swiggy_df.rating.dtype
```

The column `rating` is converted from object to float64.

#### Convert `cost` column data type from object to int.

Split the `₹` value from `cost` column and add remaining data in new column `cost_Rs`  

```python
swiggy_df['cost_Rs'] = swiggy_df.cost.str.split('₹', n=1, expand=True)[1]   # Assigning new splitted value to new column.
swiggy_df
```

Split `₹` from `cost` column and store remaining number value in new column named as `cost_Rs`

```python
swiggy_df['cost_Rs'] = swiggy_df.cost_Rs.astype(int)     # Conversion of data-type from object to int64
swiggy_df.cost_Rs.dtype
```

The column `cost_Rs` is converted from object to int64.

```python
swiggy_df.info()
```

### 3. rating

The `rating` column is the rating of the restaurant.

```python
data = swiggy_info.groupby('rating', as_index= False).agg('count')[['rating', 'name']]  # Grouping dataframe using 'rating'

df = pd.DataFrame()                                                                     # Creating an empty "df" dataframe
df = {
     '<1.0' : data[data['rating'] <= 1.0].name.sum(),
    '1.1-2.0' : data[(data['rating'] > 1.0) & (data['rating'] <= 2.0)].name.sum(),
    '2.1-3.0' : data[(data['rating'] > 2.0) & (data['rating'] <= 3.0)].name.sum(),
    '3.1-4.0' : data[(data['rating'] > 3.0) & (data['rating'] <= 4.0)].name.sum(),
    '4.1-5.0' : data[(data['rating'] > 4.0) & (data['rating'] <= 5.0)].name.sum()
}

# Storing the dictionary data in "df" dataframe and setting columns as Ratings, Restaurant_Counts
df = pd.DataFrame((df.keys(), df.values()), index=['Ratings', 'Restaurant_Counts', ]).T   
df
```

### 4. cuisine

The `cuisine` column is Cuisines/Menu that restaurant serves.

```python
swiggy_info.cuisine.unique()
```

The data in cuisine column is obtained in pair like `['Beverages,Pizzas']['Sweets, Bakery]['Biryani, Thali'] & etc.` Instead of pair, we can obtain single unique cuisine value like `Bevarages, Pizza, Sweets... and etc`.

```python
df={}                                        # Create an empty dictionary

for c in (swiggy_info['cuisine']):          # For loop used for getting list of data from cuisine column.
    l = c.split(',')                        # Split the cuisine column data using ","

    for item in l:                          # For loop used to pass single list of value in item
      if item in df:                        # Check df contains same value of cusine in "item"
        df[item] +=1                        # If yes, increment the value
      else: 
        df[item]=1                          # If no, then set new unique cuisine value to 1

# Create df as new pandas datarframe and set columns as Cusine, Restaurant Count
df = pd.DataFrame(list(zip(df.keys(), df.values())),columns =['Cuisine', 'Restaurant_Counts'])    
df          
```

```python
df_sorting = df.sort_values(by='Restaurant_Counts', ascending=False).head(20)     # Sorting the Restaurant_Count in descending order and display top 20

fig = px.bar(df_sorting, x='Cuisine', y='Restaurant_Counts',                      # Plotting the bar plot by importing plotly.express library
             template='plotly_dark', 
             color='Cuisine',
             title='Top 15 Most popular Cuisine served'
             )
fig.show()
```

In this plot, we can see top 15 most popular cuisine served by the restaurants. 
- Chinese is the most popular cuisine that are available with restaurant in swiggy across India.
- North Indian, Indian and Snacks are the followed cuisine that are available with restaurant in swiggy across India.

```python
# Create new column as cuisine_type.
swiggy_info['cuisine_type'] = swiggy_info.cuisine.str.split(',', n=1, expand=True)[0]  #Split the cuisine pair list into unique cuisine list.
```

The data in cuisine column is obtained in pair like `['Beverages,Pizzas']['Sweets, Bakery]['Biryani, Thali'] & etc`. Instead of pair, we can obtain single unique cuisine value like `Bevarages, Pizza, Sweets... and etc`.

```python
# 1. For Fast Food 

# Filtering dataframe as 'cuisine_type == Fast Food' and 'rating_count == 100+ rating' 
fast_food = swiggy_info[(swiggy_info['cuisine_type'] == 'Fast Food') & (swiggy_info['rating_count']=='100+ ratings')]   
fast_df = fast_food.groupby('cities', as_index=False).agg('count')[['cities', 'name']]                          # Grouping the dataframe with cities
ff_cities = fast_df.sort_values(by='name', ascending=False)                                                     # Sorting dataframe in descending order


# 2. For Chinese

# Filtering dataframe as 'cuisine_type == Chinese' and 'rating_count == 100+ rating' 
chinese = swiggy_info[(swiggy_info['cuisine_type'] == 'Chinese') & (swiggy_info['rating_count']=='100+ ratings')]       
chinese_df = chinese.groupby('cities', as_index=False).agg('count')[['cities', 'name']]                         # Grouping the dataframe with cities
ch_cities = chinese_df.sort_values(by='name', ascending=False)                                                  # Sorting dataframe in descending order

```

```python
fig, axes = plt.subplots(1,2, sharex=True, figsize=(25, 6))                             

# PLOT 1 For Fast Food 
ax = sns.barplot(ax=axes[0], data=ch_cities.head(10), x='name', y='cities')              # Create barplot by importing seaborn library
ax.set(xlabel = 'Total Restaurant Count where customer gave 100+ ratings ', 
       ylabel = 'Cities', 
       title = 'Top 10 cities where customer like to order Chinese');

# PLOT 2 For Chinese
ax = sns.barplot(ax=axes[1], data=ff_cities.head(10), x='name', y='cities')              # Create barplot by importing seaborn library
ax.set(xlabel = 'Total Restaurant Count where customer gave 100+ ratings ', 
       ylabel = 'Cities', 
       title = 'Top 10 cities where customer like to order Fast Food.');
```

In both plots, we can see that Bangalore, Delhi, Kolkata, Chennai and Pune are the common top 5 cities where customer like to order Chinese and Fast Food.
